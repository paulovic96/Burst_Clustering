# Preprocess Spike Data

```python
data_dir = "data/raw_data/daily_spontanous_dense/culture2_over_days/"
spike_files = [x for x in os.listdir(data_dir) if x.endswith(".spike")]
```

```python
spike_files
```

## 1. Burst Detection

```python
wagenaar_dataset.data_preprocessing(spike_files, data_dir)
```

```python
data_burst_batches = [x for x in os.listdir(data_dir) if x.find("burst_data_batch_2_2") >= 0 and x.find("tiny") <0]
tiny_burst_indices = [x for x in os.listdir(data_dir) if x.find("burst_data_batch_tiny_index_2_2") >= 0]
data_batch_names = [x.split('.')[0] for x in data_burst_batches]
data_batch_names_for_tiny_indices = [x.split('.')[0] for x in tiny_burst_indices]

data_burst_batches = wagenaar_dataset.load_batch_files_with_number_bursts(data_dir, data_batch_names)
data, culture_counts = wagenaar_dataset.merge_data_batches_ordered(data_burst_batches, day_wise = True)
culture_dict = wagenaar_dataset.get_culture_dict(culture_counts)

#tiny_burst_counts, tiny_bursts_indices = load_tiny_indices_per_culture(data_dir, data_batch_names_for_tiny_indices, day_wise = False)
#tiny_bursts_in_data_indices = get_tiny_burst_indices_for_merged_data(culture_counts, tiny_bursts_indices)

# For 2 different days 
#data_burst_batches_DayA = wagenaar_dataset.load_batch_files_with_number_bursts(data_dir_dayA, data_batch_namesA)
#data_burst_batches_DayB = wagenaar_dataset.load_batch_files_with_number_bursts(data_dir_dayB, data_batch_namesB)

#data_burst_batches_DayAB = {**data_burst_batches_DayA, **data_burst_batches_DayB}
#data_DayAB, culture_counts_DayAB = wagenaar_dataset.merge_data_batches_ordered(data_burst_batches_DayAB, day_wise = True)
#culture_dictAB = wagenaar_dataset.get_culture_dict(culture_counts_DayAB)
```

## 2. Padding and Peak alignment

```python
padded_data, data_center = wagenaar_dataset.burst_batch_padding(data, padding = "peak")
```

## 3. Channel Averaging

```python
data_burst_by_time = np.mean(padded_data,axis = 1).T
print("Burst data Batch: ", padded_data.shape)
print("Averaged over channels: ", data_burst_by_time.shape)
print("Centered at: ", data_center)

#np.save(data_dir + 'padded_data_day_21.npy', padded_data)
#np.save(data_dir + 'data_burst_by_time_culture_2_2.npy', data_burst_by_time)
```

# Plot some Examples

```python
#tiny_bursts = data_burst_by_time.T[tiny_bursts_in_data_indices]
#no_tiny_bursts = np.delete(data_burst_by_time.T,tiny_bursts_in_data_indices,axis = 0)

np.random.seed(2)
#np.random.shuffle(tiny_bursts)
np.random.shuffle(data_burst_by_time.T)

plt.figure(figsize=(20, 10))
for burst in data_burst_by_time.T[0:500]:
    plt.plot(burst)
plt.xlabel("Time (10ms bins)", fontsize=15, labelpad=10)
plt.xticks(fontsize=12)
plt.yticks(fontsize=15)
plt.ylabel("Spike Count", fontsize=15, labelpad=10)
plt.title("Day 21 Burst Examples averaged over Channels", fontsize=20, pad=20)
# plt.xlim((7500,11000))
plt.show()
#plt.savefig("burst_examples_day21.png")
```
